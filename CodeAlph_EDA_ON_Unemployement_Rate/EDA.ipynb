The date is in object type, need to convert into datetime.

```python
BeforeCovid[' Date'] = pd.to_datetime(BeforeCovid[' Date'])
```

```python
BeforeCovid.info()
```

```python
BeforeCovid = BeforeCovid.sort_values(by=' Date')
```

```python
print(BeforeCovid.columns)
```

```python
BeforeCovid.columns = BeforeCovid.columns.str.strip()
print(BeforeCovid.columns)
```

```python
unemployment_trend = BeforeCovid.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
print(unemployment_trend)
```

```python
unemployment_trend.plot()
```

### Unemployment Trend Analysis (2019â€“2020)

The unemployment rate remained relatively stable between May 2019 and February 2020, fluctuating between approximately 9% and 10%.

Starting in March 2020, a noticeable increase in unemployment was observed, rising to 10.7%. This increase became significantly more pronounced in April and May 2020, where unemployment surged to 23.6% and peaked at 24.8%, respectively.

This sharp spike coincides with the nationwide lockdown imposed due to the Covid-19 pandemic, which led to widespread business closures and job losses.

By June 2020, unemployment dropped to 11.9%, indicating the beginning of economic recovery as restrictions were gradually lifted.


```python
employment_trend = BeforeCovid.groupby('Date')['Estimated Employed'].mean()
print(employment_trend)
```

```python
employment_trend.plot()
```

### Employment Trend Analysis

The number of employed individuals remained relatively stable between May 2019 and February 2020, averaging around 7.2 to 7.7 million.

Starting in March 2020, employment began to decline, followed by a sharp drop in April 2020 to approximately 5.28 million. This significant decrease corresponds with the nationwide Covid-19 lockdown, which disrupted economic activity and led to widespread job losses.

Employment remained low in May 2020 but showed clear signs of recovery by June 2020, rising back to approximately 7.38 million, indicating gradual economic reopening and job restoration.


```python
labour_part_rate = BeforeCovid.groupby('Date')['Estimated Labour Participation Rate (%)'].mean()
print(labour_part_rate)
```

```python
labour_part_rate.plot()
```

### Labour Participation Rate Analysis

The labour participation rate remained stable between May 2019 and February 2020, averaging around 43% to 44%, indicating consistent workforce engagement during the pre-Covid period.

Starting in March 2020, the labour participation rate began to decline, followed by a sharp drop in April 2020 to approximately 35.14%, the lowest point in the observed period. This significant decrease coincides with the nationwide Covid-19 lockdown, during which many individuals were unable to work or actively seek employment due to movement restrictions and economic shutdown.

The participation rate showed partial recovery in May 2020 and continued to improve in June 2020, reaching approximately 40.55%, suggesting gradual reopening of economic activities and workforce reintegration.

### Covid-19 Period Unemployment Analysis (2020)

The unemployment rate remained relatively stable in January and February 2020, averaging around 9.2%. Beginning in March 2020, unemployment increased to approximately 10.78%, marking the initial economic impact of the Covid-19 outbreak.

A sharp spike occurred in April and May 2020, where unemployment rose to 22.23% and peaked at 23.24%, respectively. This dramatic increase corresponds directly with the nationwide lockdown, which disrupted business operations and resulted in widespread job losses.

From June 2020 onward, the unemployment rate showed significant recovery, dropping to approximately 10.91%, and continued stabilizing between 8% and 10% through October 2020. This trend reflects gradual reopening of economic activities and adaptation to pandemic conditions.

### Employment Trend During Covid-19 Period (2020)

The number of employed individuals remained high in January and February 2020, averaging approximately 15.6 million. Starting in March 2020, employment began to decline, followed by a sharp drop in April 2020 to approximately 10.57 million, representing the lowest employment level during the observed period.

This significant decline corresponds directly with the nationwide Covid-19 lockdown, which disrupted business operations and resulted in widespread job losses.

From May 2020 onward, employment steadily recovered, reaching approximately 14.5 million by September and October 2020. This recovery reflects gradual reopening of economic activities and adaptation to pandemic conditions.

### Labour Participation Rate During Covid-19 Period (2020)

The labour participation rate remained high in January and February 2020, averaging around 44%, indicating strong workforce engagement prior to the lockdown.

A sharp decline occurred in April 2020, where the participation rate dropped significantly to approximately 35.30%, the lowest point in the observed period. This decline reflects the immediate impact of the nationwide Covid-19 lockdown, during which many individuals were unable to work or actively participate in the labour force.

From May 2020 onward, the labour participation rate showed steady recovery, reaching approximately 42% by July and August 2020. This recovery and stabilization indicate gradual reopening of economic activities and workforce reintegration following the easing of lockdown restrictions.
